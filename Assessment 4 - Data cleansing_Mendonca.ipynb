```python
import pandas as pd
import numpy as np
df = pd.read_csv('Company Sales Dataset.csv')
df.head(7306)
```

```python
# Drops any duplicate rows in the dataset and saves the updated dataset
df = df.drop_duplicates()
df
```

```python
# Drops the 'Revenue' column from the dataset
df = df.drop(columns = "Revenue")
df
```

```python
# Removes the '$' sign from the revenue column and converts it to integer, then to UK currency format
df['Revenue.1'] = df['Revenue.1'].str.lstrip("$")
df['Revenue.1'] = df['Revenue.1'].astype(int)
df['Revenue.1'] = df['Revenue.1'].map("Â£{:,.0f}".format)
df
```

```python
# Converts the date column to date format
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
#Converts all data in the columns to Title case
df['Product_Category'] = df['Product_Category'].str.title()
df['Product_Name'] = df['Product_Name'].str.title()
df['Sales_Channel'] = df['Sales_Channel'].str.title()
df
```

```python
# Replaces all NaN rows with a blank value
df['Sales_Channel'] = df['Sales_Channel'].fillna('')
df
```

```python
# Removes blank cells from the Sales_Channel column
for x in df.index:
    if df.loc[x, "Sales_Channel"] == '':
        df.drop(x,inplace=True)
        
df
```

```python
# Uses Numpy to replace the NaN values in the customer_Rating column to the mean of the data in that column,
# before converting the two string columns to integer data type.
mean = df['Customer_Rating'].mean()
df['Customer_Rating'].replace(np.nan, mean, inplace=True)
df['Customer_Rating'] = df['Customer_Rating'].astype(int)
df
```

```python
# Removes outliers in the sales_amount column, where the Sales_Amount are not 
# suitable for the units_Sold.
for x in df.index:
    if (df.loc[x, 'Sales_Amount'] > 9000 and df.loc[x, 'Units_Sold'] <= 10) or (df.loc[x, 'Sales_Amount'] < 10000 and df.loc[x, 'Units_Sold'] >= 40):
        df.drop(x,inplace=True)

df   
```

```python
# Converts Sales_Amount to integer then to currency format
df['Sales_Amount'] = df['Sales_Amount'].astype(int)
df['Sales_Amount'] = df['Sales_Amount'].map("Â£{:,.0f}".format)
df
```

```python
# Goes through the dataset and standardises the categories that each product belongs too.
for x in df.index:
    if df.loc[x, 'Product_Name'] == "Laptop":
        df.loc[x, 'Product_Category'] = "Computers"
    elif df.loc[x, 'Product_Name'] == "Smartphone":
        df.loc[x, 'Product_Category'] = "Mobile Phones"
    elif df.loc[x, 'Product_Name'] == "Headphones" or df.loc[x, 'Product_Name'] == "Tablet" :
        df.loc[x, 'Product_Category'] = "Electronics"
df
```

```python
# Creates a new CSV file with all the changes
df.to_csv('CleanedData.csv')
```
